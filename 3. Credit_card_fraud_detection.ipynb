Above operation shaows that the data is highly unbalanced so we have to balance it .

```python
# seperating data for analysis

legit = credit_card_data[credit_card_data.Class == 0]

fraud = credit_card_data[credit_card_data.Class == 1]
```

Dealing with unbalanced data (undersampling)

Here we are building simple dataset which cantain same amount of fraud and ligit transcation

number of fraudlant transaction -->492

```python
legit_sample = legit.sample(n=492)
```

Concatinate 2 datasets

```python
new_data = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data["Class"].value_counts()
```

```python
new_data.groupby("Class").mean()
```

 Data cleaning and undersampling is done now its time to apply model.

```python
# Here we are seperating features and lebel..

x = new_data.drop(columns= ["Class"], axis=1)

y = new_data["Class"]
```

```python
# now apply train_test_split on dataset

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.2,
                                                    random_state=23,
                                                    stratify=y)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

Now it is time to  train the model and check accuracy

# Logistic regression

```python
# Starting with Logistic Regression

model = LogisticRegression(random_state=23)

model.fit(x_train, y_train)
```

```python
y_train_pred = model.predict(x_train)
```

```python
# train_accuracy

train_accuracy = accuracy_score(y_train_pred, y_train)
print("training_accuracy :-",train_accuracy)
```

```python
#test_accuracy

y_test_pred = model.predict(x_test)
test_accuracy = accuracy_score(y_test_pred, y_test)
print("testing_accuracy :- ",test_accuracy)
```

```python
# Confusion matrix
confusion_matrix(y_test_pred,y_test)
```
